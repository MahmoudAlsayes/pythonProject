```python
import numpy as np

import heartpy as hp
import pandas as pd
import matplotlib.pyplot as plt
with open("files/8_1.txt","r") as file:
    d6=file.read()
```

```python
d6.tolist()
```

```python
# Your string data
# data_str = '2003.0\t2003.0\t1978.0\t1978.0\t1978.0\t1993.0\t1993.0\t1986.0\t...'

# Split the string by tabs
data_list = d6.split('\t')

# Convert each item to float, then to integer, handling any conversion errors
data_int = []
for item in data_list:
    try:
        value = float(item)  # Convert to float
        data_int.append(int(value))  # Convert to integer
    except ValueError:
        # Handle the case where the conversion fails
        print(f"Warning: Could not convert '{item}' to float and integer")

# Print the result
print(data_int)
```

```python
plt.figure(figsize=(12,6))
plt.plot(data_int)
plt.show()
```

```python
import numpy as np

def moving_average(data_int, window_size):
    """Calculate the moving average of the data with a given window size."""
    return np.convolve(data_int, np.ones(window_size)/window_size, mode='valid')


# Parameters
window_size = 5  # Example window size

# Apply filter
filtered_data = moving_average(data_int, window_size)

print(filtered_data)
```

```python
filtered_data = np.array(filtered_data)
```

```python
filtered_data
```

```python
import matplotlib.pyplot as plt

# Plot original and filtered data
plt.figure(figsize=(12, 6))
plt.plot(data_int, label='Original Data', alpha=0.5)
plt.plot(np.arange(window_size-1, len(data_int)), filtered_data, label='Filtered Data', color='red')
plt.legend()
plt.show()
```

1. Mean and Standard Deviation

```python
import numpy as np

# Example PPG signal


# Compute Mean
mean_value = np.mean(data_int)

# Compute Standard Deviation
std_dev = np.std(data_int)

print("Mean:", mean_value)
print("Standard Deviation:", std_dev)
```

2. Peak-to-Peak Interval

```python
# Compute Peak-to-Peak Interval
peak_to_peak = np.max(data_int) - np.min(data_int)

print("Peak-to-Peak Interval:", peak_to_peak)
```

3. Root Mean Square (RMS)

```python
# Compute RMS
rms_value = np.sqrt(np.mean(np.square(data_int)))

print("RMS:", rms_value)
```

4. Entropy

```python
from scipy.stats import entropy

# Compute Histogram
hist, bin_edges = np.histogram(data_int, bins='auto')

# Compute Entropy
entropy_value = entropy(hist)

print("Entropy:", entropy_value)
```

5. Autocorrelation


```python
from numpy.fft import fft, ifft

def autocorrelation(signal):
    result = ifft(fft(signal) * np.conj(fft(signal)))
    return np.real(result)

# Compute Autocorrelation
auto_corr = autocorrelation(data_int)

print("Autocorrelation:", auto_corr)
```

6. Frequency Domain Features (e.g., Power Spectral Density)


```python
from scipy.signal import welch

# Compute Power Spectral Density
frequencies, psd = welch(data_int)

print("Frequencies:", frequencies)
print("Power Spectral Density:", psd)
```

7. Number of Peaks


```python
from scipy.signal import find_peaks

# Find Peaks
peaks, _ = find_peaks(data_int)

# Number of Peaks
num_peaks = len(peaks)

print("Number of Peaks:", num_peaks)
```
