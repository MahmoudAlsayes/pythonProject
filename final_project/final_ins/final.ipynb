```python
import numpy as np

import heartpy as hp
import pandas as pd
import matplotlib.pyplot as plt
with open("files/8_1.txt","r") as file:
    d6=file.read()
```

```python
# Your string data
# data_str = '2003.0\t2003.0\t1978.0\t1978.0\t1978.0\t1993.0\t1993.0\t1986.0\t...'

# Split the string by tabs
data_list = d6.split('\t')

# Convert each item to float, then to integer, handling any conversion errors
data_int = []
for item in data_list:
    try:
        value = float(item)  # Convert to float
        data_int.append(int(value))  # Convert to integer
    except ValueError:
        # Handle the case where the conversion fails
        print(f"Warning: Could not convert '{item}' to float and integer")

# Print the result
print(data_int)
```

```python
import numpy as np


# Compute Mean
mean_value = np.mean(data_int)

# Compute Standard Deviation
std_dev = np.std(data_int)

print("Mean:", mean_value)
print("Standard Deviation:", std_dev)


##
# Compute Peak-to-Peak Interval
peak_to_peak = np.max(data_int) - np.min(data_int)

print("Peak-to-Peak Interval:", peak_to_peak)
##

# Compute RMS
rms_value = np.sqrt(np.mean(np.square(data_int)))

print("RMS:", rms_value)


## 

from scipy.stats import entropy

# Compute Histogram
hist, bin_edges = np.histogram(data_int, bins='auto')

# Compute Entropy
entropy_value = entropy(hist)

print("Entropy:", entropy_value)


###

from scipy.signal import find_peaks

# Find Peaks
peaks, _ = find_peaks(data_int)

# Number of Peaks
num_peaks = len(peaks)

print("Number of Peaks:", num_peaks)


## 
from scipy.stats import skew, kurtosis

# Compute Skewness
skewness = skew(data_int)

# Compute Kurtosis
kurtosis_value = kurtosis(data_int)

print("Skewness:", skewness)
print("Kurtosis:", kurtosis_value)
result={
    "Mean:": mean_value,
"Standard Deviation:" :std_dev,
"Peak-to-Peak Interval:": peak_to_peak,
"RMS:": rms_value,

"Entropy:": entropy_value,
"Number of Peaks:": num_peaks,
"Skewness:": skewness,
"Kurtosis:": kurtosis_value,
}
```

```python
import os
import numpy as np
import pandas as pd
from scipy.stats import entropy, skew, kurtosis
from scipy.signal import find_peaks

# Initialize the DataFrame
df = pd.DataFrame(columns=["Mean", "Standard Deviation", "Peak-to-Peak Interval", "RMS", "Entropy", "Number of Peaks", "Skewness", "Kurtosis"])

base_dir = 'files/'  # Replace with the actual directory path where your files are stored

for i in range(1, 420):  # Loop from 1 to 219 (inclusive)
    filename = f'{i}_1.txt'  # Adjust the filename pattern as per your actual file names
    filepath = os.path.join(base_dir, filename)
    
    # Check if the file exists before attempting to read it
    if os.path.exists(filepath):
        # Load file content and convert it to a numpy array
        file_content = np.loadtxt(filepath)  # Assuming the file content is numeric
        
        # Compute Mean
        mean_value = np.mean(file_content)
        
        # Compute Standard Deviation
        std_dev = np.std(file_content)
        
        # Compute Peak-to-Peak Interval
        peak_to_peak = np.max(file_content) - np.min(file_content)
        
        # Compute RMS
        rms_value = np.sqrt(np.mean(np.square(file_content)))
        
        # Compute Histogram
        hist, bin_edges = np.histogram(file_content, bins='auto')
        
        # Compute Entropy
        entropy_value = entropy(hist)
        
        # Find Peaks
        peaks, _ = find_peaks(file_content)
        
        # Number of Peaks
        num_peaks = len(peaks)
        
        # Compute Skewness
        skewness = skew(file_content)
        
        # Compute Kurtosis
        kurtosis_value = kurtosis(file_content)
        
        # Append results to DataFrame
        result = {
            "Mean": mean_value,
            "Standard Deviation": std_dev,
            "Peak-to-Peak Interval": peak_to_peak,
            "RMS": rms_value,
            "Entropy": entropy_value,
            "Number of Peaks": num_peaks,
            "Skewness": skewness,
            "Kurtosis": kurtosis_value
        }
        
        df = df._append(result, ignore_index=True)
                
    else:
        print(f"File '{filename}' not found.")

# Optionally save the DataFrame to a CSV file
df.to_csv('results.csv', index=False)
```
