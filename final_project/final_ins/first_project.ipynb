```python
SP_presure = 140
DP_presure = 90
# 120/80

def classify_blood_pressure(sp,dp):
    if sp>SP_presure or dp>DP_presure:
        return 1
    else:
        return 0

data["blood_pressure"]=data.apply(lambda x:classify_blood_pressure(x["Systolic Blood Pressure(mmHg)"],x["Diastolic Blood Pressure(mmHg)"]),axis=1)
```

```python
data.drop(["Systolic Blood Pressure(mmHg)","Diastolic Blood Pressure(mmHg)"],axis=1,inplace=True)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
x= data.drop("blood_pressure", axis=1)
y=data["blood_pressure"]
smote=SMOTE(random_state=42)
x_smote, y_smote=smote.fit_resample(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote,test_size=0.3, random_state=42)
```

# SVM 

```python
from  sklearn.svm import SVC

svm=SVC(kernel="linear",random_state=42,probability=True)
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
# Predict and evaluate
from sklearn.metrics import classification_report ,confusion_matrix
svm_predictions = svm.predict(x_test)
print("SVM Classification Report:")
print(classification_report(y_test, svm_predictions))

print("SVM Confusion Matrix:")
print(confusion_matrix(y_test, svm_predictions))
```
